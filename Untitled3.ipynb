```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud,STOPWORDS
```

```python
df=pd.read_csv('top_50.csv', encoding='cp1252')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'Track.Name':'Track', 'Artist.Name':'Artist'}, inplace=True)
df.info()
```

```python
plt.figure(figsize=(10,10))
string=str(df.Artist)
wordcloud=WordCloud(stopwords=STOPWORDS,
                   background_color='red',
                   height=1000,
                   width=1000).generate(string)
plt.imshow(wordcloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10,10))
string=str(df.Genre)
wordcloud=WordCloud(stopwords=STOPWORDS,
                    background_color='black',
                   height=1000,
                   width=1000).generate(string)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show
```

```python
plt.figure(figsize=(10,10))
plt.title('Correlation')
x=sns.heatmap(df.corr(),
           linewidth=2.6,
           annot=True,
              cbar=False,
           center=0)
```

```python
sns.catplot(y = "Genre", kind = "count",
            palette = "pastel", edgecolor = ".6",
            data = df)
```
